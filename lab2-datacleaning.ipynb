# Lab 2 - Data Cleaning

## Missing Data

```python
# reading in the data
import pandas as pd
import numpy as np
Location = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(Location, header = None)
df.columns = headers
df.head()
```

```python
# convert "?" to nan
df.replace("?", np.nan, inplace=True)
df.head()
```

```python
# identifying missing data
# isnull()
missing_data = df.isnull()
missing_data.head()
```

```python
# notnull()
present_data = df.notnull()
present_data.head()
```

```python
# count missing values in each column - using simple for loop
for c in missing_data.columns.values.tolist():
    print(c)
    print (missing_data[c].value_counts())
    print(" ")
    # true - represents the number of missing data points
```

```python
# replace by mean:
# normalized-losses
df["normalized-losses"].replace(np.nan, 
                        df["normalized-losses"].astype("float").mean(), 
                        inplace=True)
# stroke
df["stroke"].replace(np.nan,
                    df['stroke'].astype("float").mean(),
                    inplace=True)
# bore
df["bore"].replace(np.nan,
                  df["bore"].astype("float").mean(),
                  inplace=True)
# horsepower
df["horsepower"].replace(np.nan,
                        df["horsepower"].astype("float").mean(),
                        inplace=True)
# peak-rpm
df["peak-rpm"].replace(np.nan,
                      df["peak-rpm"].astype("float").mean(),
                      inplace=True)

df.head()
```

```python
# replace by frequency:
# num-of-doors
df["num-of-doors"].replace(np.nan,"four",inplace=True)
```

```python
# drop entire row
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True,inplace=True)
df.head()
```

## Correct Data Format

```python
# checking data types
df.dtypes
```

```python
# converting columns to the correct data types
df[["bore","stroke"]] = df[["bore","stroke"]].astype("float")
df["normalized-losses"] = df["normalized-losses"].astype("int")
df[["price","peak-rpm"]] = df[["price","peak-rpm"]].astype("float")
df.dtypes
```

## Data Standardization

```python
# transform mpg to L/100km
# formula for unit conversion - L/100km = 235 / mpg
df.head()
```

```python
# city mpg
df["city-mpg"] = 235 / df["city-mpg"]
df["highway-mpg"] = 235 / df["highway-mpg"]
df.rename(columns={"city-mpg":"city L/100km",
                  "highway-mpg":"highway L/100km"},
                  inplace = True)
df.head()
```

## Data Normalization

```python
# scaling the columns length, width, and height
# objective - normalize these values so that their value ranges from 0-1
# using simple-feature scaling
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df["height"] = df["height"]/df["height"].max()
```

## Binning Data

```python
# split the variable horsepower into 3 discrete bins (low, medium, high)
# check to make sure variable is the right type 
df.dtypes # is an object - will need to convert it to integer
```

```python
df["horsepower"] = df["horsepower"].astype("int", copy=True)
```

```python
# plot horsepower with a histagram to examine its distribution
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(df["horsepower"])
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Bins")
```

```python
# finding bins of equal distance
bins = np.linspace(df["horsepower"].min(),df["horsepower"].max(),4)
names = ["low","medium","high"]
df["horsepower-bins"] = pd.cut(df["horsepower"],
                              bins,
                              labels=names,
                              include_lowest=True)
df[["horsepower","horsepower-bins"]].head(10)
```

```python
# displaying the number of vehicles in each bin
df["horsepower-bins"].value_counts()
```

```python
# creating a bargraph of our newly binned data
%matplotlib inline
plt.bar(x=names,height=df["horsepower-bins"].value_counts())
# need to give plt.bar what should be on the x and y axis
plt.xlabel("Horsepower Category")
plt.ylabel("Counts")
plt.title("Horsepower by Category")
```

## Dummy Variables

We want to convert categorical variables to numeric variables so that we can include them in later regression/modeling.

```python
# converting fuel type into dummy variables
df.columns
```

```python
# create the dummy variables
dummy_variable = pd.get_dummies(df["fuel-type"])
dummy_variable.head()
```

```python
# insert the dummy variables back into the dataframe
df = pd.concat([df,dummy_variable],axis=1)
df.head()
```

```python
# drop original column fuel-type from the dataframe
df.drop("fuel-type",axis=1, inplace=True)
```

```python
df.head()
```

```python
# create a dummy variable for aspiration
dummies = pd.get_dummies(df["aspiration"])
# change names for clarity
dummies.rename(columns={"std":"aspiration-std","turbo":"aspiration-turbo"},
              inplace=True)
dummies.head()
```

```python
# join together with existing dataframe
df = pd.concat([df,dummies],axis=1)
# remove the column aspiration
df.drop("aspiration",axis=1,inplace=True)
df.head()
```

```python
# output the final csv
df.to_csv("auto-clean.csv")
```
