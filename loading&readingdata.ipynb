```python
# Saving Data to SQL

# creating the dataframe
import pandas as pd
names = ["Bob","Jessica","Mary","John", "Mel"]
grades = [76,95,77,78,99]
GradeList = list(zip(names,grades))
df = pd.DataFrame(data=GradeList,
                 columns=["Names","Grades"])
df
```

```python
# exporting data to SQL
import os
import sqlite3 as lite
db_filename = r'mydb.db'
con = lite.connect(db_filename)
df.to_sql("mytable",
         con,
         schema=None,
          if_exists="replace",
          index=True,
          index_label=None,
          chunksize=None,
          dtype=None)
con.close()
#df.to_sql?
```

```python
# Random Numbers and Creating Random Data
# creating a random list of baby names
import pandas as pd
from numpy import random
from numpy.random import randint
names = ["Bob","Jessica","Mary","John","Mel"]
# set the seed
random.seed(500)
# generate a random number between 0 and the length of names
n = randint(low=0,high=len(names))
print(names[n]) # selects the name corresponding to the random index
# selecting 1000 random names and filling in array with them
randnames = []
for i in range(1000):
    name = names[randint(low=0,high=len(names))]
    randnames.append(name)
    
```

```python
# selecting 1000 random numbers
births = []
for i in range(1000):
    births.append(randint(low=0,high=1000))
```

```python
# merge the two random lists together into a dataframe
BabyDataSet2 = list(zip(randnames,births))
df = pd.DataFrame(data=BabyDataSet2, columns = ["Names","Births"])
df

# export the dataframe to excel
writer = pd.ExcelWriter("RandBirths.xlsx", engine="xlsxwriter")
df.to_excel(writer,sheet_name="Sheet1")
writer.save()
```

```python
# Practive with Random Numbers
import pandas as pd
from numpy import random
from numpy.random import randint
names = ["Jon","Sherry","Martha","Sarah","Laura"]
randnames = []
for i in range(250):
    name = names[randint(low=0,high=len(names))]
    randnames.append(name)
parkingtix = []
for i in range(250):
    parkingtix.append(randint(low=0,high=25))

ParkingData2 = list(zip(randnames,parkingtix))
df = pd.DataFrame(data=ParkingData2,columns=["Names","Tickets"])
df.to_csv("parkingtickets.csv", index=False, header=True)
```
